# **Correlation**
It is a measure of dependencies or association between two variables i.e how does one variable changes with respect to other
Correlation refers how close two variables are to have a linear relationship with each other

There are three types of correlation:-
1. Positive Correlation: if One variable increases then other variable also increases
2. Negative Correlation: If one variable increases then other variable decreases 
3. Zero Correlation: Change in one variable doesn't affect other variable
**The value of correlation always lies between -1 to +1**

# **Correlation between independent variable**

STRONG CORRELATION: if correlation>0.55 it is termed as strong correlation

Why we need to find correlation between independent variables

*  Two independent variables that are highly correlated with each other and also highly correlated with the dependent variable then using both the variables in the model may result in suboptimal or poorly understood model
*   The model will most likely use one of those highly correlation variable in the equation and may consider the other variable as statistically insignificant variable
