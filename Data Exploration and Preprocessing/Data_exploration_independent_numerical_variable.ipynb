#**Imputing**

1.   Object variable imputation

# **Continous Variable imputation**
In this we will be imputing all the numeric variables

```python
# combining all the numeric variable under a single list
number_columns=['No of Bathrooms','Flat Area (in Sqft)', 'Lot Area (in Sqft)','Area of the House from Basement (in Sqft)','Latitude','Longitude','Living Area after Renovation (in Sqft)']
```

for imputation we will bw using impute from sklearn

```python
import sklearn 
from sklearn.impute import SimpleImputer
imputer= SimpleImputer(strategy='median')

# missing_value specifies how missing values are represented in our database by default it is np.nan
# strategy specifies how we impute the missing value
df[number_columns]=imputer.fit_transform(df[number_columns])
```

```python
df.info()
```

There are two steps in fit_transform function:-

1. FITTING PHASE: in this phase the function calculate the median value w.r.t every column that we passed as a parameter and stores it

2. TRANSFORM PHASE: in this phase function does the actual action of locating the missing value and imputing it using median strategy

# **Object or Categorical Variable imputing**
In this we will be using mode strategy

```python
imputer=SimpleImputer(strategy='most_frequent')
df['Zipcode'].shape
#as the column of zipcode have only row detail and column detail are missing so we need to add it
```

```python
column=df['Zipcode'].values.reshape(-1,1)
column.shape
df['Zipcode']=imputer.fit_transform(column)
df.info()
```

#**To convert numerical variable to categorical variables**

```python
df['Zipcode']=df['Zipcode'].astype(object)
df.info()
```

# **To convert a categorical variable to numerical variable**
The steps are as fallow:
1. Find out all the unique value in the categorical variable
2. replace the unique value with their corresponding numerical values

```python
df['No of Times Visited'].unique()
```

```python
# recoding them into numerical values using pandas.map function

mapping={
    'None':'0',
    'Once':'1',
    'Twice':'2',
    'Thrice':'3',
    'Four':'4'
}

#using pandas.map function pass mapping into it
df['No of Times Visited']=df['No of Times Visited'].map(mapping)

#checking the unique value
df['No of Times Visited'].unique()
```

3. Another reason for variable transformation is to derive a new variable from existing variable or combining two variable to transform a variable so that it become more useful

   In the below codes we will be deriving two new variable from the variable 'Renovated Years'


*   Ever Renovated: this will be assigned 'yes' if the renovated year have non zero value otherwise it will be assigned 'no;
*   Year Since Renovation: if renovated year have non zero value then it will calculated the year since the house is last renovated



```python
# in this we will be using where function from numpy
import numpy as np
df['Ever Renovated']= np.where(df['Renovated Year']==0,'No','Yes')
```

```python
df['Purchase Year']=pd.DatetimeIndex(df['Date House was Sold']).year
df['Year Since Renovation']=np.where(df['Ever Renovated']=='Yes',abs(df['Purchase Year']-df['Renovated Year']),0)
```
